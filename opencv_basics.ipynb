```python
from cv2 import cv2
from matplotlib import pyplot as plt
```

#Plotting


```python
img=cv2.imread("resources/face1.jpg",0)
```

```python
plt.imshow(img)
```

```python
img1=cv2.imread("resources/face9.jpg",0)
width = int(225)
height = int(225)
dim = (width, height)
resized = cv2.resize(img1, dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized)
```

```python
plt.imshow(result)
```
